### Music Mood? Genre? Classifier Testing

## Step 1: Import Libraries
We begin by importing the necessary libraries, which include pandas for data manipulation, numpy for numerical operations, and matplotlib and seaborn for data visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

## Step 2: Load the Data
We load the data from a CSV file (`data.csv`). This dataset will be used for analysis in the following steps.

```python
# load and inspect the dataset
data_path = 'data.csv'  
data = pd.read_csv(data_path)
```

## Step 3: Data Overview
In this section, we take a quick look at the data using `head()` and `info()` methods to understand its structure, column types, and any missing values.

```python
# basic dataset information
print("Dataset Information:\n")
print(data.info())
print("\nSample Rows:\n", data.head())
```

## Step 4: Data Cleaning
Here, we check for missing values and handle them appropriately. This is crucial to ensure data integrity for the analysis.

```python
# select features and labels
features = data.drop(columns = ['filename', 'label'])
labels = data['label']
```

## Step 5: Exploratory Data Analysis (EDA)
We analyze the data through descriptive statistics and visualizations to uncover patterns, correlations, and insights.

```python
# feature engineering: scale features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# dimensionality reduction with PCA
pca = PCA(n_components = 2)
pca_features = pca.fit_transform(scaled_features)
```

```python
# plot PCA
plt.figure(figsize=(8, 6))
sns.scatterplot(x = pca_features[:, 0], y = pca_features[:, 1], hue = labels, palette = "Set1")
plt.title("PCA Visualization of Music Data")
plt.xlabel("Principal Component 1") # Principal Component 1: Represents the direction of maximum variance in the data, capturing the most important patterns.
plt.ylabel("Principal Component 2") # Principal Component 2: Orthogonal to PC1, it captures the second most significant variance, highlighting additional data structure.
plt.legend(title = "Genre", loc = 'upper right')
plt.show()
```

This plot reveals the relationships between different music genres based on key musical features. Some genres are tightly grouped (indicating similarity), while others are more isolated (indicating uniqueness). PCA helps us see these patterns in two dimensions, making complex, multidimensional data more understandable.

```python
# split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(scaled_features, labels, test_size = 0.2, random_state = 42)
```

```python
# initializes and trains a Random Forest Classifier
# training a Random Forest Classifier to leverage its ability to handle non-linear relationships, identify important features, and provide robust performance for mood classification.
rf = RandomForestClassifier(random_state = 42)
rf.fit(X_train, y_train)
```

```python
# predictions and evaluation
y_pred = rf.predict(X_test)
```

```python
# displays the confusion matrix to visualize the performance of the classifier by showing the counts of true positive, true negative, false positive, and false negative predictions.
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))

# prints the classification report to provide detailed metrics such as precision, recall, and F1-score for each class, giving a deeper understanding of model performance.
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# calculates and display the overall accuracy of the model, which is the proportion of correctly predicted samples out of the total.
print("\nAccuracy:", accuracy_score(y_test, y_pred))
```

This report evaluates the performance of a classification model trained to predict music genres. The confusion matrix summarizes how often each genre was correctly predicted and where the model made errors, showing patterns of misclassification. The classification report provides precision, recall, and F1-score metrics for each genre, highlighting the model's strengths (e.g., "classical" and "metal") and weaknesses (e.g., "reggae" and "country"). Finally, the accuracy of 56% reflects the proportion of correct predictions overall, indicating room for improvement in distinguishing certain genres.

```python
# gets and ranks feature importance scores from the Random Forest model in descending order.
# feature importance visualization
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1] # feature importance visualization
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]
```

```python
# plots a bar chart to visualize the importance of each feature as determined by the Random Forest model.
plt.figure(figsize = (10, 6))
plt.title("Feature Importances")
plt.bar(range(features.shape[1]), importances[indices], align = "center")
plt.xticks(range(features.shape[1]), features.columns[indices], rotation = 90)
plt.grid(color='gray', linewidth=0.5, alpha=0.5)  # lighter grid lines
plt.tight_layout()
plt.show()
```

This plot shows the importance of various features in predicting the target variable, music genre (label). Features like chroma_stft and rmse, represented by taller bars, are highly influential, as they capture key audio patterns that help differentiate genres. In contrast, features like bass and tempo contribute less, indicating lower distinctiveness or correlation with other features.

The visualization highlights the models priorities, aiding feature selection by identifying the most relevant attributes. This can simplify the model while still maintaining it's accuracy and offers insights into which audio characteristics are most likely to be indicative of genre classifications.

```python
# performs hyperparameter tuning for the Random Forest model using GridSearchCV to optimize performance.
# hyperparameter tuning
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, cv = 3, scoring = 'accuracy', n_jobs = -1)
grid_search.fit(X_train, y_train)
```

`GridSearchCV` helps find the best settings for a `RandomForestClassifier` by testing different combinations of parameters, using 3-fold cross-validation to check accuracy. The final model uses 200 trees (`n_estimators=200`) for reliable and accurate predictions.

```python
print("\nBest Parameters:\n", grid_search.best_params_)
print("\nBest Grid Search Accuracy:", grid_search.best_score_)
```

The best parameters found by GridSearchCV were `max_depth=None` (no limit on tree depth), `min_samples_leaf=1` (leaf nodes can have as few as 1 sample), `min_samples_split=2` (a node can split with at least 2 samples), and `n_estimators=200` (200 trees in the forest). These settings achieved an accuracy of 65.25%, balancing model complexity to improve predictions.

```python
# retrains with best parameters
best_rf = grid_search.best_estimator_
y_pred_best = best_rf.predict(X_test)
```

```python
print("\nFinal Confusion Matrix:\n", confusion_matrix(y_test, y_pred_best))
print("\nFinal Classification Report:\n", classification_report(y_test, y_pred_best))
print("\nFinal Accuracy:", accuracy_score(y_test, y_pred_best))
```

The final confusion matrix shows how well the model predicted each music genre. It achieved an overall accuracy of 60%, with the highest precision for disco (0.82) and the lowest precision for hip hop (0.42). Classical had the highest recall (1.00), while disco and reggae had the lowest recall (0.43), indicating that the model performed best for disco in terms of precision, but struggled the most with recall for disco. The F1-score provides a balanced measure of precision and recall, with classical achieving the highest F1-score (0.87) due to its perfect recall, while genres like disco and reggae had lower F1-scores due to the trade-offs between precision and recall. Support indicates the number of true instances for each genre, helping provide context to these scores.

### Visualizing Feature Correlations with a Heatmap
To better understand the relationships between features in our dataset, we'll use a heatmap. 
This visualization highlights correlations, where values close to 1 or -1 indicate strong relationships between variables.


```python
# filters numeric columns only
numeric_data = data.select_dtypes(include=['float64', 'int64']) # selects only the numeric columns (integers and floating-point numbers) from the `data` DataFrame to account for the error `ValueError: could not convert string to float: 'blues.00081.au'`, which occurs when non-numeric data is present in the correlation calculation.

# calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# set up the matplotlib figure
plt.figure(figsize=(14, 10))  # Increase the figure size for better clarity

# draw the heatmap with Seaborn
sns.heatmap(
    correlation_matrix, 
    annot=True, 
    fmt='.2f', 
    cmap='coolwarm', 
    cbar=True, 
    square=True, 
    annot_kws={"size": 8},  # adjust font size of annotations
    linewidths=0.5,  # add space between cells
    linecolor='black'  # add visible borders to separate cells
)

# add title
plt.title('Feature Correlation Heatmap', fontsize=16)
plt.tight_layout()  # adjust layout to prevent clipping
plt.show()
```

The first cluster, forming a tightly packed region, includes features such as spectral_centroid, spectral_bandwidth, and rolloff, which are known to be interrelated measures of spectral properties. This group displayed intense correlations (dark red blocks), indicating strong linear relationships among these features. The second cluster, less tightly packed but still significant, included features such as mfcc11, mfcc13, and other lower and higher-order MFCCs. These observations suggest that some MFCCs capture overlapping information, which might influence their individual contributions to the model.
